```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

from sklearn.model_selection import train_test_split
from sklearn.pipeline import make_pipeline
```

```python
kredit = pd.read_csv('tmdb_5000_credits.csv', index_col = 'movie_id')
#baca database kredit
```

```python
movies = pd.read_csv('tmdb_5000_movies.csv', index_col = 'id')

#merge yang sama

data = pd.merge(movies,kredit)
print(data.shape)

data = data.loc[data['revenue']!=0]
data['revenue'].dropna(inplace=True)
print(data.shape)

X_train,X_test,y_train,y_test = train_test_split(data.drop(['revenue'],axis = 1), data['revenue'])

exploratory = X_train.copy()
#kita gunakan kopian supaya dataset asli tidak berantakan
```

```python
#Cek missing value
missing_percent = exploratory.isna().mean()*100
missing_count = exploratory.isna().sum()
pd.concat([missing_percent.round().rename('Persentase Missing'), missing_count.rename()], axis = 1)
```

```python
drop_kolom = ['original_title', 'overview', 'title', 'tagline']

exploratory = exploratory.drop(drop_kolom, axis = 1)
```
